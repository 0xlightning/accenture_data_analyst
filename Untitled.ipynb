```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../Reactions-Copy1.xlsx')
df
```

```python
df['Category'] = df['Category'].str.lower()
```

```python
total_scores = df.groupby('Category')['Score'].sum()
total_scores
```

```python
top_five_categories = total_scores.sort_values(ascending=False).head(5)[::-1]
top_five_categories
```

```python
plt.title('Top 5 Category by Aggregate Score "Popularity"')
plt.xlabel('Aggregate Score')
plt.ylabel('CATEGORIES')
plt.barh(top_five_categories.index, top_five_categories)
plt.xlim(60000, 78000) ;
```

```python
plt.title('Popularity percentage share for Top 5 Categories')
plt.pie(top_five_categories, labels=top_five_categories.index,autopct="%0.1f%%", startangle=180, shadow=True, explode=[0.04,0,0,0.06,0.1]);
```
